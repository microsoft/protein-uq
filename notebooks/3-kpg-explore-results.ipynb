```python
def visualize_results(csv):

    df = pd.read_csv(results_dir+'/'+csv)

    df['y_test'].hist(label='y_test', alpha=0.3)
#     plt.xlabel('y_test')
#     plt.show()
    df['preds_mean'].hist(label='preds_mean', alpha=0.3)
    plt.legend()
#     plt.xlabel('preds_mean')
    plt.show()

#     df['preds_std'].hist()
#     plt.xlabel('preds_std')
#     plt.show()

    x, y = df['y_test'], df['preds_mean']
    axmin, axmax = min(min(x), min(y))*0.9, max(max(x), max(y))*1.1
    plt.scatter(x, y, alpha=0.1)
    plt.xlabel('y_test')
    plt.ylabel('preds_mean')
    plt.xlim((axmin, axmax))
    plt.ylim((axmin, axmax))
    plt.show()
    
    return
```

```python
split_dict = {
    'aav_1': 'des_mut' ,
    'aav_2': 'mut_des',
    'aav_3': 'one_vs_many',
    'aav_4': 'two_vs_many',
    'aav_5': 'seven_vs_many',
    'aav_6': 'low_vs_high',
    'aav_7': 'sampled',
    'meltome_1' : 'mixed_split',
    'meltome_2' : 'human',
    'meltome_3' : 'human_cell',
    'gb1_1': 'one_vs_rest',
    'gb1_2': 'two_vs_rest',
    'gb1_3': 'three_vs_rest',
    'gb1_4': 'sampled',
    'gb1_5': 'low_vs_high',
    #     
    'des_mut': 'des_mut' ,
    'mut_des': 'mut_des',
    'one_vs_many': 'one_vs_many',
    'two_vs_many': 'two_vs_many',
    'seven_vs_many': 'seven_vs_many',
    'low_vs_high': 'low_vs_high',
    'sampled': 'sampled',
    'mixed_split' : 'mixed_split',
    'human' : 'human',
    'human_cell' : 'human_cell',
    'one_vs_rest': 'one_vs_rest',
    'two_vs_rest': 'two_vs_rest',
    'three_vs_rest': 'three_vs_rest',
    'sampled': 'sampled',
    'low_vs_high': 'low_vs_high',
    'meltome_mixed_short999': 'meltome_mixed_short999',
}

results_df['split'] = results_df['split'].map(split_dict)
results_df
```

```python
gb1_df = results_df.loc[results_df['dataset']=='gb1']
aav_df = results_df.loc[results_df['dataset']=='aav']
meltome_df = results_df.loc[results_df['dataset']=='meltome']
```

### Coverage vs. RMSE

```python
colors = ['tab:blue', 'tab:orange', 'tab:green', 'tab:red']
shapes = ['o','^','s','v','+','x']

for landscape, landscape_df in zip(['gb1','aav','meltome'], [gb1_df, aav_df, meltome_df]):
    for i, split in enumerate(set(landscape_df['split'])):
        df_ = landscape_df.loc[landscape_df['split']==split]
        for j, model in enumerate(set(df_['model'])):
            df__ = df_.loc[df_['model']==model]
            plt.scatter(df__['test rmse'], 
                        df__['percent coverage'], 
                        c=colors[i],
                        marker=shapes[j],
                        label=f'{split} - {model}')
    plt.xlabel('Test RMSE')
    plt.ylabel('Coverage (95%)')
    plt.title(landscape)
    plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
    plt.show()
```
