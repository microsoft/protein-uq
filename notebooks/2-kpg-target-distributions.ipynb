```python
import pandas as pd
```

```python
def visualize_range(csv):
    df = pd.read_csv(csv)
    print(df['target'].max() - df['target'].min())
    df['target'].hist()
    return
```

```python
visualize_range('../data/aav/splits/mut_des.csv')
```
