```python
def visualize_results(csv):

    df = pd.read_csv(results_dir+'/'+csv)

    df['y_test'].hist(label='y_test', alpha=0.3)
#     plt.xlabel('y_test')
#     plt.show()
    df['preds_mean'].hist(label='preds_mean', alpha=0.3)
    plt.legend()
#     plt.xlabel('preds_mean')
    plt.show()

#     df['preds_std'].hist()
#     plt.xlabel('preds_std')
#     plt.show()

    x, y = df['y_test'], df['preds_mean']
    axmin, axmax = min(min(x), min(y))*0.9, max(max(x), max(y))*1.1
    plt.scatter(x, y, alpha=0.1)
    plt.xlabel('y_test')
    plt.ylabel('preds_mean')
    plt.xlim((axmin, axmax))
    plt.ylim((axmin, axmax))
    plt.show()
    
    return
```

### Coverage vs. RMSE

```python
def plot_two_vars(x,y):
    
    palette = {
        "Sampled": "tab:blue",
        "Mixed": "tab:blue", 
        "Mut-Des": "tab:pink",
        "7 vs. Many": "tab:green",
        "3 vs. Rest": "tab:purple",
        "2 vs. Rest": "tab:orange",
        "1 vs. Rest": "tab:red",
    }
    
    for landscape, landscape_df_ in zip(['gb1','aav','meltome'], [gb1_df, aav_df, meltome_df]):
        landscape_df = landscape_df_.sort_values(['split_rank','Model'])
        sns.scatterplot(data=landscape_df,
                        x=x,
                        y=y,
#                         hue='Split',
#                         style='Model',
                        hue='Model',
                        style='Split',
#                         palette=palette,
                        s=100,
                        alpha=0.8)
        
        if x == "test rmse":
            plt.xlabel("Test RMSE", fontsize=15)
        elif x == "average width / range":
            plt.xlabel("Average Width / Range", fontsize=15)
        else:
            plt.xlabel(x, fontsize=15)
        
        if y == "percent coverage":
            plt.ylabel("Percent Coverage", fontsize=15)
        elif y == "miscalibration area":
            plt.ylabel("Miscalibration Area", fontsize=15)
        else:
            plt.xlabel(y, fontsize=15)
            
        if landscape == "gb1":
            plt.title("GB1", fontsize=15)
        elif landscape == "aav":
            plt.title("AAV", fontsize=15)
        elif landscape == "meltome":
            plt.title("Meltome", fontsize=15)
            
        plt.legend(bbox_to_anchor=(1.05, 1.0), loc='upper left')
        
        if x == "test rmse":
            x_ = "rmse"
        elif x == "average width / range":
            x_ = "width"
        else:
            x_ = x
        
        if y == "percent coverage":
            y_ = "coverage"
            plt.ylim((-0.05,1.05))
        elif y == "miscalibration area":
            y_ = "area"
            plt.ylim((-0.05,0.55))
        else:
            y_ = y
        
        plt.xticks(fontsize=12)
        plt.yticks(fontsize=12)
        
        plt.savefig(f"{landscape}_{y_}_{x_}.pdf", bbox_inches="tight")
        
        plt.show()
        
    return
```

```python
plot_two_vars('average width / range', 'percent coverage')
```

```python
plot_two_vars('test rmse', 'miscalibration area')
```
